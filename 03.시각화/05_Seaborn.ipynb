## Seaborn 데이터 시각화

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

## 1. 카운트 플롯

```python
sns.countplot(x='class', data=titanic)
plt.title('타이타닉호의 각 클래스별 승객 수')
plt.show()
```

```python
sns.countplot(x='day', data=tips)
plt.title('요일별 팁 준 횟수')
plt.show()
```

## 2. 다차원 실수형 데이터

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02)
plt.show()
```

```python
sns.pairplot(iris)
plt.title("Iris Data의 Pair Plot")
plt.show()
```

```python
sns.pairplot(iris, hue='species', markers=['o', 's', 'D']);
```

## 3. 2차원 카테고리 데이터
- heatmap

```python
tt = titanic.pivot_table(index='class', columns='sex', aggfunc='size')
tt
```

```python
sns.heatmap(tt, annot=True, fmt='d', cmap=sns.light_palette('magenta'))
plt.title('Titanic 선실별 남녀 승객수')
plt.show()
```

```python
flights_passengers = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(10,10))
plt.title("연도, 월 별 승객수에 대한 Heatmap")
sns.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1)
plt.show()
```

## 4. 다차원 복합 데이터
- barplot

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("요일 별, 성별 전체 팁의 Histogram")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 성별 전체 요금의 평균")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 성별 전체 요금의 평균")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue='sex',data=tips)
plt.title("요일 별, 성별 전체 요금의 평균")
plt.show()
```
